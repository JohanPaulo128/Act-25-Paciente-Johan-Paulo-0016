###Crear la clase Paciente, con 3 atributos (nombre, edad, diagnóstico), 2 funciones leerdatos(), mostrardatos() y 2 objetos (paciente1, paciente2).






![infoclasepaciente999.png](https://raw.githubusercontent.com/JohanPaulo128/Act-25-Paciente-Johan-Paulo-0016/refs/heads/main/infoclasepaciente999.png)

```python
class Paciente:
    def __init__(self):
        # Atributos inicializados en vacío
        self.nombre = ""
        self.edad = 0
        self.diagnostico = ""

    # Función 1: Leer datos (Interactiva con input e int)
    def leer_datos(self, numero_paciente):
        print(f"\n--- Ingresando datos del Paciente {numero_paciente} ---")
        self.nombre = input("Nombre del paciente: ")
        self.edad = int(input("Edad del paciente: "))  # Uso de int()
        self.diagnostico = input("Diagnóstico médico: ")

    # Función 2: Mostrar datos
    def mostrar_datos(self):
        print("\n--- Ficha Médica ---")
        print(f"Nombre: {self.nombre}")
        print(f"Edad: {self.edad} años")
        print(f"Diagnóstico actual: {self.diagnostico}")
        print("--------------------")

    # Función 3: Acción adicional
    def realizar_checkup(self):
        print(f"Mensaje: El paciente {self.nombre} está pasando a revisión de signos vitales.")

# --- Lógica Principal del Programa ---

# Creación de Objeto 1
paciente1 = Paciente()
paciente1.leer_datos(1)

# Creación de Objeto 2
paciente2 = Paciente()
paciente2.leer_datos(2)

# Ejecución de funciones de salida
print("\n" + "="*30)
print("RESULTADOS DE LA INVESTIGACIÓN")
print("="*30)

paciente1.mostrar_datos()
paciente1.realizar_checkup()

paciente2.mostrar_datos()
paciente2.realizar_checkup()
```
